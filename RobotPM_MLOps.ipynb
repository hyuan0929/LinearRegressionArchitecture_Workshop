# Linear Regression MLOps Workshop Project

## Project Goals
1. Detect robot work periods from raw multi-axis force telemetry
2. Aggregate each work period into mean/peak force features
3. Split work periods into detection intervals and run linear regression per interval:
   - from scratch (gradient descent)
   - scikit-learn (OLS)
4. Evaluate models (RMSE, MAE, R2) and plot regression lines
5. Design robust alert thresholds and generate experiment tracking output:
   - `experiments/results.csv`

## Folder Structure
- `data/raw/` raw data + work-period summary
- `data/preprocessed/` standardized modeling table
- `data/models/` theta tables, evaluated tables, plots, thresholds
- `experiments/` tracking outputs
- `src/` modular pipeline code
- `configs/` YAML config
